```python
import numpy as np
```

```python

all_x = np.array([13, 15, 16])

for i in [0, 1, 2]:
  print(all_x[i])
```

```python
for x in all_x:
  print(x)
```

```python
all_x = np.array([13, 15, 16, 20, 21, 25])
for i in np.arange(0, 6):
  print(all_x[i])
```

```python
for i in range(0, 6, 2):
  print(i)
```
